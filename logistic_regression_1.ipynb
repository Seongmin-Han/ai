```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Load the data
file = '/content/drive/MyDrive/Colab Notebooks/titanic.csv'
data = pd.read_csv(file)

# Drop unnecessary columns
data = data.drop(['Name', 'Ticket'], axis=1)

# Convert categorical features to numerical using Label Encoding
categorical_features = ['Sex', 'Embarked']
for feature in categorical_features:
    le = LabelEncoder()
    data[feature] = le.fit_transform(data[feature]) # Fit and transform the data to numerical labels

# Split the data into training and testing sets
X = data.drop('Survived', axis=1)
y = data['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

# Create and train the model
model = LogisticRegression(C=0.1, penalty='l2', solver='saga', max_iter=2000, random_state=100)
model.fit(X_train, y_train)

# Make predictions and evaluate the model
pred = model.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(f"Accuracy: {accuracy}")
```
