```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/Colab Notebooks/Mall_Customers.csv'
data = pd.read_csv(file_path)

X = data[['Annual Income (k$)', 'Spending Score (1-100)']]

kmeans = KMeans(n_clusters=3, random_state=0)
data['Cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(10, 7))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=data['Cluster'], cmap='viridis', marker='o', edgecolor='k', s=100)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X')
plt.title('K-Means Clustering (K=3) of Annual Income vs. Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]

inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 7))
plt.plot(K_range, inertia, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.xticks(K_range)
plt.show()
```

```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]

kmeans_3 = KMeans(n_clusters=3, random_state=0)
data['Cluster_3'] = kmeans_3.fit_predict(X)

plt.figure(figsize=(10, 7))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=data['Cluster_3'], cmap='viridis', marker='o', edgecolor='k', s=100)
plt.scatter(kmeans_3.cluster_centers_[:, 0], kmeans_3.cluster_centers_[:, 1], s=200, c='red', marker='X')
plt.title('K-Means Clustering (K=3) of Annual Income vs. Spending Score with Centroids')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()

optimal_k = 5
kmeans_optimal = KMeans(n_clusters=optimal_k, random_state=0)
data['Cluster_optimal'] = kmeans_optimal.fit_predict(X)

plt.figure(figsize=(10, 7))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=data['Cluster_optimal'], cmap='viridis', marker='o', edgecolor='k', s=100)
plt.scatter(kmeans_optimal.cluster_centers_[:, 0], kmeans_optimal.cluster_centers_[:, 1], s=200, c='blue', marker='X')
plt.title(f'K-Means Clustering (K={optimal_k}) of Annual Income vs. Spending Score with Centroids')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
